```python
# Analysis of H-angle for Normalised Pressure > 12.5kPa
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
#Loading dataset
df = pd.read_csv('Train1.csv')
df1 = pd.read_csv('Test1.csv')

# Extract features (X) and target variable (y)
X = df[['H_angle_inv', 'Seat_distance']]  # Add more features as needed
y = df['Normalised_Pressure']

X1 = df1[['H_angle_inv', 'Seat_distance']]
y1 = df1['Normalised_Pressure']

# Creating polynomial features
for degree in range(1,6):#Checking different degrees of polynomial regression from degree 1 to 5
    print("For Degree =", degree)
    
    poly = PolynomialFeatures(degree=degree)
    X_train_poly = poly.fit_transform(X)
    X_test_poly = poly.transform(X1)
    
    poly_reg = LinearRegression()
    poly_reg.fit(X_train_poly, y)


    y_train_pred = poly_reg.predict(X_train_poly)
    y_test_pred = poly_reg.predict(X_test_poly)

    # Evaluating the model
    train_rmse = mean_squared_error(y, y_train_pred, squared=False)
    test_rmse = mean_squared_error(y1, y_test_pred, squared=False) # root mean squared when H-angle and seat distance is known 

    
    print("Coefficients:", poly_reg.coef_)
    print("Intercept:", poly_reg.intercept_)
    print("Training RMSE:", train_rmse)
    print("Test RMSE:", test_rmse)
    print()

#Using the model to predict error in H_angle_calculation
X_Normalised_Pressure = df['Normalised_Pressure']
l1 = list(X_Normalised_Pressure)
X_Seat_distance = df['Seat_distance']
l2 = list(X_Seat_distance)
Y_H_angle = df['H-angle']
l3 = list(Y_H_angle)
n = len(l3)
Y_pred_train = list()
for i in range (0,n):
    #coefficients obtained from regression
    a = 182
    b = 0.0096245
    c = -1.305
    y = a/(l1[i]-b*l2[i]-c)
    Y_pred_train.append(y)
train_rmse_h_angle = mean_squared_error(l3, Y_pred_train, squared=False)
print("--------------------------------------------------------Result---------------------------------------------------------")
print("The root mean squared error in H-angle prediction when Normalised pressure and seat distance is known using our model:")
print(train_rmse_h_angle,"degrees")

```
